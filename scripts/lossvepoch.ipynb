```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
csv_file = '/global/ml4hep/spss/twamorkar/PELICAN/predict/muonlevel1_newsplit.metrics.train.csv'  # Update with the path to your CSV file
df = pd.read_csv(csv_file)

# Plot loss vs. epoch (assuming each row corresponds to an epoch)
plt.figure(figsize=(8, 6))
plt.plot(df['loss'], marker='o', linestyle='-', color='b', label='Loss')

# Add labels and title
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss vs Epoch')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['accuracy'], marker='o', linestyle='-', color='b', label='Loss')

# Add labels and title
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Epoch')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
